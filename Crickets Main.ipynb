```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

Chirps_sec = [20.0, 16.0, 19.8, 18.4, 17.1, 15.5, 14.7, 15.7, 15.4, 16.3, 15.0,17.2, 16.0, 17.0, 14.4]
Temperature = [88.6, 71.6, 93.3, 84.3, 80.6, 75.2, 69.7, 71.6, 69.4, 83.3, 79.6, 82.6, 80.6, 83.5, 76.3]
xx = np.linspace(65,95,20)

def lin_model(x, a, b):
    return a*x + b 

a,b = curve_fit(lin_model, Temperature, Chirps_sec)[0]
print (a,b)

plt.ylabel('Chirps/sec')
plt.xlabel('Temperature')
plt.title('Chirp Rate of Crickets Vs Temperature')
plt.plot(Temperature, Chirps_sec, '.')
plt.plot(xx, lin_model(xx, a, b), label='y=%sx+%s'% (round(a,3), round(b,3)))
plt.legend(loc=4)
plt.ylim([0,21])
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

time = [0, .04, .08, .12, .16, .20, .24]
distance = [0, 3.3, 5.8, 7.7, 9.1, 9.9, 10.2]
xx=np.linspace(0., .50, 20)

def lin_model(x,a,b,c):
    return a*x**2+b*x+c

a,b,c = curve_fit(lin_model, time, distance)[0]
print(a,b,c)

plt.text(.24, 10.2, 'maximum point')
plt.xlabel('Time')
plt.ylabel('Height')
plt.title('Basketball Traveling')
plt.plot(time, distance, '.', label='data')
plt.plot(xx, lin_model(xx, a, b, c), label='%st^2 + %st + %s' % (a, b, c))
plt.legend(loc=4)
plt.show()
```
