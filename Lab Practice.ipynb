```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4]),([5,6,7,8])
plt.ylabel('tv seen')
plt.xlabel('age')
plt.title('Amout of tv through the years')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4,5]), ([2,4,5,6,8]),'k+')
plt.ylabel('increase in size')
plt.xlabel('time in days')
plt.title('cell growth')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np 
from scipy.optimize import curve_fit

x = np.random.uniform(0.,100.,100)
y = 3. * x + 2. + np.random.normal(0.,10.,100)
def line(x, a, b):
    return a*x + b
xm = np.linspace(0.,100.,100)
lin_vars, crap = curve_fit(line, x, y)
plt.plot(line(xm, lin_vars[0], lin_vars[1]), 'k-^')
plt.plot(x, y, '.')
```
